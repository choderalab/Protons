# Initial calibration of amino acids
Viewing the SAMS calibration plots for the titratable amino acids.
* Total number of steps requested = 100000, many stopped before that.
* 'burn-in' length = 66666
* b = 0.5
* NCMC steps at 1fs = 10000 (10ps)
* MD steps per iteration = 500

```python
import pickle
import numpy as np
from matplotlib import pylab as plt
%matplotlib inline
```

## Loading data from NCMC calibration of TYR

```python
deviation, weights, work, times = pickle.load(open('tyr/prev_out.pickle','r'))

plt.plot(range(len(deviation)), deviation)
plt.xlabel('Iteration')
plt.ylabel('Absolute deviation from target probability')
plt.title('NCMC calibration of TYR in explicit solvent')
plt.show()

g_k = [w[1] for w in weights]

plt.plot(range(len(g_k)), g_k)
plt.xlabel('Iteration')
plt.ylabel('2nd SAMS weight')
plt.title('NCMC calibration of TYR in explicit solvent')
plt.show()
```
